```python
import numpy as np 
import pandas as pd 
import plotly.graph_objects as go
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
world = pd.read_csv("file:///C:/Users/javig/AppData/Local/Temp/owid-covid-data.csv")
uk = world[world["iso_code"]=="GBR"]
port = world[world["iso_code"]=="PRT"]
ven = world[world["iso_code"]=="VEN"]
turk = world[world["iso_code"]=="TUR"]
esp = world[world["iso_code"]=="ESP"]
world.columns
```

```python
uk = uk.drop(uk.columns.difference(["iso_code","date",'total_cases',"total_deaths", 'total_deaths_per_million']), 1)
port = port.drop(port.columns.difference(["iso_code","date",'total_cases',"total_deaths", 'total_deaths_per_million']), 1)
ven = ven.drop(ven.columns.difference(["iso_code","date",'total_cases',"total_deaths", 'total_deaths_per_million']), 1)
turk = turk.drop(turk.columns.difference(["iso_code","date",'total_cases',"total_deaths", 'total_deaths_per_million']), 1)
esp = esp.drop(esp.columns.difference(["iso_code","date",'total_cases',"total_deaths", 'total_deaths_per_million']), 1)
```

```python
uk
```

```python
uk.dropna( inplace= True)
port.dropna( inplace= True)
ven.dropna( inplace= True)
turk.dropna( inplace= True)
esp.dropna( inplace= True)
```

```python
import plotly.express as px 
```

```python
uk_graph= uk.drop(uk.columns.difference(["date",'total_cases',"total_deaths"]), 1)
port_graph= port.drop(port.columns.difference(["date",'total_cases',"total_deaths"]), 1)
ven_graph=ven.drop(ven.columns.difference(["date",'total_cases',"total_deaths"]), 1)
turk_graph=turk.drop(turk.columns.difference(["date",'total_cases',"total_deaths"]), 1)
esp_graph=esp.drop(esp.columns.difference(["date",'total_cases',"total_deaths"]), 1)
```
